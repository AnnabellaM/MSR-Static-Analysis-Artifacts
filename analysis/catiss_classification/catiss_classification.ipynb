```python
from simpletransformers.classification import ClassificationModel

# Create a ClassificationModel
model = ClassificationModel('roberta', './catiss/model', use_cuda=False)
```

```python
import pandas as pd
import sys
sys.path.append('../../')
from utils.dataloader import get_issues

# 0 - bug
# 1 - enhancement
# 2 - question

# Load the dataset into a pandas dataframe.
issues = get_issues()

title_body = []

for index, issue in issues.iterrows():
    title_body.append(issue['title'] + ' ' + str(issue['body']))

predictions = model.predict(title_body)

title_body = pd.DataFrame(title_body, columns=['text'])

title_body['prop:category'] = predictions[0]
title_body['prop:category'].map({0: 'bug', 1: 'enhancement', 2: 'question'}, inplace=True)
```

```python
issues = issues[['id', 'repo', 'host', 'prop:state', 'prop:category', 'prop:resolution', 'prop:comments', 'prop:users', 'prop:files', 'prop:loc']]
issues.to_csv('../../data/issues_properties.csv', index=False)
```
