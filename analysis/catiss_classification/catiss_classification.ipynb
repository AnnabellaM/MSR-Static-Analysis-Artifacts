```python
from simpletransformers.classification import ClassificationModel

# Create a ClassificationModel
model = ClassificationModel('roberta', './model', use_cuda=False)
```

```python
import pandas as pd
import sys
sys.path.append('../../../')
from utils.dataloader import get_issues

# 0 - bug
# 1 - enhancement
# 2 - question

# Load the dataset into a pandas dataframe.
issues = get_issues()

title_body = []

for issue in issues:
    title_body.append(issue['title'] + ' ' + str(issue['body']))

predictions = model.predict(title_body)

title_body = pd.DataFrame(title_body, columns=['text'])

title_body['predictions'] = predictions[0]
```

```python
issues_df = pd.DataFrame(issues)
title_body['id'] = issues_df['id']
title_body['repo'] = issues_df['repo']
title_body['host'] = issues_df['host']
title_body['url'] = issues_df['html_url']
title_body['state'] = issues_df['state']
title_body['created_at'] = issues_df['created_at']

title_body.to_csv('../../results/csv/catiss/catiss_predictions.csv', index=False)
```
