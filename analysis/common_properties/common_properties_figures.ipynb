```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sys
sys.path.append('../../')

from utils.dataloader import get_issues

sns.set(style='whitegrid')

issues = get_issues(filter="properties")
```

```python
# Open vs. Closed Issues
status_counts = issues['prop:state'].value_counts()
plt.figure(figsize=(8, 8))

patches, texts, autotexts = plt.pie(
    status_counts,
    labels=['Closed', 'Open'],
    pctdistance=0.7,
    labeldistance=0.5,
    autopct='%1.1f%%',
    colors=['#aec7e8', '#ffbb78'],  # lighter blue and orange
    startangle=100,
)

# Set font size for labels
for text in texts:
    text.set_fontsize(40)  # Adjust label font size
    text.set_fontweight('bold')  # Make label bold
    text.set_horizontalalignment('center')  # Align label to center

# Set font size for percentages (autotexts)
for autotext in autotexts:
    autotext.set_fontsize(60)  # Adjust percentage font size

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
# increase font size
plt.tight_layout()
plt.savefig('../../results/figures/common_properties/state.pdf', format='pdf', dpi=300)
plt.show()
```

```python
# issue categories distribution
category_counts = issues['prop:category'].value_counts()
plt.figure(figsize=(8, 8))

patches, texts, autotexts = plt.pie(
    category_counts,
    labels=['Bug', 'Enhan..', 'Ques.'],
    autopct='%1.1f%%',
    colors=['#aec7e8', '#ffbb78', '#98df8a'],  # lighter blue, orange, green
    startangle=140,
    pctdistance=0.7,
    labeldistance=0.5,
)

# Set font size for labels
for text in texts:
    text.set_fontsize(40)  # Adjust label font size
    text.set_fontweight('bold')  # Make label bold
    text.set_horizontalalignment('center')  # Align label to center

# Set font size for percentages (autotexts)
for autotext in autotexts:
    autotext.set_fontsize(60)  # Adjust percentage font size

plt.axis('equal')
plt.tight_layout()
plt.savefig('../../results/figures/common_properties/category.pdf', format='pdf', dpi=300)
plt.show()
```

```python
# Resolution Time
resolution_times = issues[issues['prop:state'] == 'closed']['prop:resolution']

plt.figure(figsize=(12, 6))
plt.hist(resolution_times, bins=50, color='#1f77b4', alpha=0.7)
plt.gca().set_xlim(left=0)
plt.yscale('log')
plt.xlabel('Resolution Time (days)', fontsize=30)
plt.ylabel('Count (log scale)', fontsize=30)
plt.xticks(fontsize=35, rotation=45)
plt.yticks(fontsize=35)
plt.tight_layout()
plt.savefig('../../results/figures/common_properties/resolution.pdf', format='pdf', dpi=350)
plt.show()
```
