```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sys
sys.path.append('../../')

from utils.dataloader import get_issues

sns.set(style='whitegrid')

issues = get_issues()
```

```python
# Open vs. Closed Issues
status_counts = issues['prop:state'].value_counts()
plt.figure(figsize=(8, 8))

patches, texts, autotexts = plt.pie(
        status_counts,
        labels=status_counts.index,
        pctdistance=0.7,
        labeldistance=1.25,
        autopct='%1.1f%%',
        colors=['#1f77b4', '#ff7f0e'],
        startangle=90,
        textprops={'fontsize': 40}
)

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
# increase font size
plt.tight_layout()
plt.savefig('../../results/figures/common_properties/all_repos/state.pdf', format='pdf', dpi=300)
plt.show()
```

```python
# issue categories distribution
category_counts = issues['prop:category'].value_counts()
plt.figure(figsize=(8, 8))

patches, texts, autotexts = plt.pie(
        category_counts,
        labels=category_counts.index,
        autopct='%1.1f%%',
        colors=['#1f77b4', '#ff7f0e', '#2ca02c'],
        startangle=150,
        textprops={'fontsize': 40},
        pctdistance=0.7,
        labeldistance=1.1
)

plt.axis('equal')
plt.tight_layout()
plt.savefig('../../results/figures/common_properties/all_repos/category.pdf', format='pdf', dpi=300)
plt.show()
```

```python
# Resolution Time
resolution_times = issues[issues['prop:state'] == 'closed']['prop:resolution']

plt.figure(figsize=(12, 6))
plt.hist(resolution_times, bins=50, color='#1f77b4', alpha=0.7)
plt.yscale('log')
plt.xlabel('Resolution Time (days)', fontsize=30)
plt.ylabel('Count (log scale)', fontsize=30)
plt.xticks(fontsize=35, rotation=45)
plt.yticks(fontsize=35)
plt.tight_layout()
plt.savefig('../../results/figures/common_properties/all_repos/resolution.pdf', format='pdf', dpi=350)
plt.show()
```

```python
# comment count
plt.figure(figsize=(12, 6))
plt.hist(issues['prop:comments'], bins=range(0, issues['prop:comments'].max() + 1), color='#ff7f0e', alpha=0.7)
plt.yscale('log')
plt.xlabel('# of Comments', fontsize=30)
plt.ylabel('Count (log scale)', fontsize=30)
plt.xticks(fontsize=35, rotation=45)
plt.yticks(fontsize=35)
plt.tight_layout()
plt.savefig('../../results/figures/common_properties/all_repos/comments.pdf', format='pdf', dpi=350)
plt.show()
```

```python
# user participation count
plt.figure(figsize=(12, 6))
plt.hist(issues['prop:users'], bins=range(0, issues['prop:users'].max() + 1), color='#1f77b4', alpha=0.7)
plt.yscale('log')
plt.xlabel('# of Unique Users', fontsize=30)
plt.ylabel('Count (log scale)', fontsize=30)
plt.xticks(fontsize=35, rotation=45)
plt.yticks(fontsize=35)
plt.tight_layout()
plt.savefig('../../results/figures/common_properties/all_repos/users.pdf', format='pdf', dpi=350)
plt.show()
```

```python
# files changed
plt.figure(figsize=(12, 6))
plt.hist(issues['prop:files'], bins=50, color='#2ca02c', alpha=0.7)
plt.yscale('log')
plt.xlabel('# of Files Changed (100)', fontsize=30)
plt.ylabel('Count (log scale)', fontsize=30)

xticks = plt.xticks()[0]
plt.xticks([x for x in xticks if int(x/100) != -5], [int(x/100) for x in xticks if int(x/100) != -5], fontsize=35, rotation=45)
plt.yticks(fontsize=35)
plt.tight_layout()
plt.savefig('../../results/figures/common_properties/all_repos/files.pdf', format='pdf', dpi=350)
plt.show()
```

```python
# loc changed
plt.figure(figsize=(12, 6))
plt.hist(issues['prop:loc'], bins=50, color='#d62728', alpha=0.7)
plt.yscale('log')
plt.xlabel('# of LoC Changed (100,000)', fontsize=30)
plt.ylabel('Count (log scale)', fontsize=30)
xticks = plt.xticks()[0]
plt.xticks([x for x in xticks if int(x/100000) != -1], [int(x/100000) for x in xticks if int(x/100000) != -1], fontsize=35, rotation=45)
plt.yticks(fontsize=35)
plt.tight_layout()
plt.savefig('../../results/figures/common_properties/all_repos/loc.pdf', format='pdf', dpi=350)
plt.show()
```
